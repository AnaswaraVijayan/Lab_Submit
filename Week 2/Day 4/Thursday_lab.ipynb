```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
customer_analysis=pd.read_csv("Data_Marketing_Customer_Analysis_Round3.csv")
customer_analysis['sales_channel'].unique()
```

```python
customer_numeric=customer_analysis.select_dtypes(include=np.number)
customer_categorical=customer_analysis.select_dtypes(include=object)
customer_numeric.info()
customer_categorical.info()
```

```python
#Show a plot of the total number of responses.
cs=customer_analysis['response'].value_counts()
cs
```

```python
sns.barplot(x=cs.index,y=cs)
plt.xlabel("response")
plt.ylabel("count")
plt.title("A plot for the total number of response")
```

```python
#Show a plot of the response by the sales channel.
sc=customer_analysis.groupby(['sales_channel'])['response'].count().reset_index()
sc
```

```python
sns.countplot(data=sc, x='response', hue='sales_channel')
```

```python
#Show a plot of the response by the total claim amount.
tc=customer_analysis.groupby(['total_claim_amount'])['response'].count().reset_index()
tc
```

```python
sns.boxplot(data=tc, x='response', y='total_claim_amount')
```
